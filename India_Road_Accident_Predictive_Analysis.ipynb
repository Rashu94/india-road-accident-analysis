###### Q1-calculate the Accident Hotspot Risk Index (AHRI)
formula :AHRI = (Fatalities × 0.6) + (Casualties × 0.4) + (Accidents × 0.2) 
       - (TrafficControl × 0.1)

```python
# Group data by City and sum relevant accident factors
city_accidents = df.groupby("City Name").agg({
    "Number of Fatalities": "sum",
    "Number of Casualties": "sum",
    "State Name": "count",  # Count total accidents
    "Traffic Control Presence": "sum"
}).rename(columns={"State Name": "Total Accidents"})
```

```python
# Group data by City and sum relevant accident factors
# Recalculate AHRI using a weighted formula
city_accidents["AHRI"] = (
    city_accidents["Number of Fatalities"] * 0.6 + 
    city_accidents["Number of Casualties"] * 0.3 + 
    city_accidents["Total Accidents"] * 0.1
)

print(city_accidents["AHRI"].describe())  # Check new AHRI values
print(city_accidents["AHRI"].head(10))    # Ensure first 10 rows are nonzero

```

```python
city_accidents["AHRI"] = city_accidents["AHRI"] * 100  # Scale up if needed
```

```python
city_accidents["AHRI"] = city_accidents["AHRI"].clip(lower=0)  # Convert negatives to zero
```

```python
print(city_accidents["AHRI"].describe())
```

```python
# Select top 10 most dangerous cities
top_cities = city_accidents.head(20)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_cities.index, y=top_cities["AHRI"], palette="Reds",legend=False)

plt.xticks(rotation=45)
plt.xlabel("City Name")
plt.ylabel("Accident Hotspot Risk Index (AHRI)")
plt.title(" Top 10 Most Dangerous Cities Based on AHRI ")

plt.show()
```

```python

# Capitalize month names if needed
df["Month"] = df["Month"].str.capitalize()

# Set correct month order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Group by Month and sum casualties
monthly_trend = df.groupby("Month")["Number of Casualties"].sum().reindex(month_order)

# Plot the trend
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_trend.index, y=monthly_trend.values, marker="o", color="blue", linewidth=3)

plt.title("Seasonal Road Safety Model – Monthly Casualty Trends", fontsize=16, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Casualties", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

```

```python


# Capitalize months for consistency
df["Month"] = df["Month"].str.capitalize()

# Ensure month order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Convert 'Month' to categorical type for ordered plotting
df["Month"] = pd.Categorical(df["Month"], categories=month_order, ordered=True)

# Set the visual theme
sns.set_theme(style="whitegrid")

# Create the barplot
plt.figure(figsize=(12, 8))
sns.barplot(
    data=df,
    x="Month",
    y="Number of Casualties",
    hue="Year",
    estimator="sum",
    palette="Spectral"
)

plt.title("📊 Monthly Accident Casualties by Year", fontsize=16, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Casualties", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Year")
plt.tight_layout()
plt.show()
```

# Prediction  for 2025 

```python
yearly_fatalities["Rolling_Avg"] = yearly_fatalities["Number of Fatalities"].rolling(window=3).mean()
predicted_2025 = int(yearly_fatalities["Rolling_Avg"].iloc[-1])  # Take last rolling average
print(f"🔮 Predicted Fatalities in 2025 (3-Year Rolling Average): {predicted_2025}")
```

# 2025  Prediction Chart 

```python
# Append prediction for 2025
predicted_df = pd.DataFrame({"Year": [2025], "Number of Fatalities": [predicted_2025]})
yearly_fatalities_extended = pd.concat([yearly_fatalities, predicted_df], ignore_index=True)

# Plot with forecast
plt.figure(figsize=(10, 5))
sns.lineplot(data=yearly_fatalities_extended, x="Year", y="Number of Fatalities", marker="o", color="teal")
plt.axvline(x=2024.5, linestyle="--", color="gray")
plt.text(2025, predicted_2025, f"Prediction: {predicted_2025}", color="red", weight="bold")
plt.title("📈 Fatalities Forecast Including 2025")
plt.ylabel("Number of Fatalities")
plt.tight_layout()
plt.show()
```
